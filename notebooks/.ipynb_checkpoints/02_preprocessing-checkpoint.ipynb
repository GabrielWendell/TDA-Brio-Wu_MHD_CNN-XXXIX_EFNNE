## 1. Normalization

### 1.1 Mathematical Model

We'll use **global $z$-score normalization per variable**.

For each scalar field $f\in\{\rho,p\}$:
$$
\mu_{f}=\frac{1}{N_{t}N_{x}}\sum_{k=0}^{N_{t}-1}\sum_{i=0}^{N_{x}-1}f_{k}\left(i\right)\quad,\quad\sigma_{f}^{2}=\frac{1}{N_{t}N_{x}}\sum_{k,i}\left[f_{k}\left(i\right)-\mu_{f}\right]^{2}\text{ },
$$
and we define
$$
\tilde{f}_{k}\left(i\right)=\frac{f_{k}\left(i\right)-\mu_{f}}{\sigma_{f}}\text{ }.
$$

This has nice properties:
- Inputs to the neural networks are approximately **zero-mean and unit-variance**, which stabilizes optimization.
- TDA is invariant under translation and scale in many constructions, but using normalized fields helps when we later compare different variables or experiments.

We'll also **save** the means and standard deviations so taht we can go back to physical units later.

### 1.2 Compute and store normalized fields

````python
# Global means
mean_rho = rho.mean()
mean_p   = p.mean()

# Global standard deviations
std_rho = rho.std()
std_p   = p.std()

print(f"mean_rho = {mean_rho:.6f}, std_rho = {std_rho:.6f}")
print(f"mean_p   = {mean_p:.6f}, std_p   = {std_p:.6f}")

# Z-score normalization
rho_norm = (rho - mean_rho) / std_rho
p_norm   = (p   - mean_p)   / std_p

print("rho_norm shape:", rho_norm.shape)
print("p_norm shape:", p_norm.shape)

print("\n\nCheck normalized means/stds:")
print("rho_norm mean, std:", rho_norm.mean(), rho_norm.std())
print("p_norm  mean, std:", p_norm.mean(),  p_norm.std())
````

## 2. Time-pair construction for ML

We now construct the dataset of input–target pairs for the **one-step temporal predictor**:

- For each $k=0,\ldots,N_{t}-2$:
> - **Input**
$$
X_{k}\left(x\right)=\left(\begin{array}{c}
\tilde{\rho}\left(t_{k},x\right)\\
\tilde{p}\left(t_{k},x\right)
\end{array}\right)\in\mathbb{R}^{2\times N_{x}}\text{ }.
$$
> - **Target**
$$
Y_{k}\left(x\right)=\left(\begin{array}{c}
\tilde{\rho}\left(t_{k+1},x\right)\\
\tilde{p}\left(t_{k+1},x\right)
\end{array}\right)\text{ }.
$$


Stacking all pairs, we get tensors:
$$
X\in\mathbb{R}^{\left(N_{t}-1\right)\times2\times N_{x}}\quad,\quad Y\in\mathbb{R}^{\left(N_{t}-1\right)\times2\times N_{x}}\text{ }.
$$
This structure is natural for a 1D CNN (`channels = 2`, `length = N_x`). If later you want a fully-connected network, we can simply flatten the last two axes.

### 2.1 Build X, Y in "channel-first" format

````python
# Number of time steps and spatial points
N_t, N_x = rho_norm.shape
assert N_t == len(times)

# Inputs: states at t_k, k = 0,...,N_t-2
# Targets: states at t_{k+1}
X = np.stack([rho_norm[:-1], p_norm[:-1]], axis = 1)  # shape (N_t-1, 2, N_x)
Y = np.stack([rho_norm[1:],  p_norm[1:]],  axis = 1)  # shape (N_t-1, 2, N_x)

print("X shape:", X.shape)
print("Y shape:", Y.shape)
````

### 2.2 Flattened version for fully-connected models

````python
X_flat = X.reshape(X.shape[0], -1)  # (N_t-1, 2*N_x)
Y_flat = Y.reshape(Y.shape[0], -1)  # (N_t-1, 2*N_x)

print("X_flat shape:", X_flat.shape)
print("Y_flat shape:", Y_flat.shape)
````

## 3. Save processed tensors to disk

We now save everything into a single `.npz` file for reproducibility and quick loading from other notebooks (`02_tda_ground_truth`, `03_models_baseline`, etc.).

We'll store:
- `times`, `x`
- `rho`, `p` (raw)
- `rho_norm`, `p_norm`
- `mean_rho`, `std_rho`, `mean_p`, `std_p`
- `X`, `Y` (and optionally `X_flat`, `Y_flat`)
- Extra fields (if you want them later)

Save to `data/processed/brio_wu_tensors.npz`

````python
processed_dir = "../data/processed"
os.makedirs(processed_dir, exist_ok = True)

npz_path = os.path.join(processed_dir, "brio_wu_tensors.npz")

save_dict = {
    "times": times,
    "x": x,
    "rho": rho,
    "p": p,
    "rho_norm": rho_norm,
    "p_norm": p_norm,
    "mean_rho": mean_rho,
    "std_rho": std_rho,
    "mean_p": mean_p,
    "std_p": std_p,
    "X": X,
    "Y": Y,
    "X_flat": X_flat,
    "Y_flat": Y_flat,
}

np.savez_compressed(npz_path, **save_dict)
print(f"Saved processed tensors to {npz_path}")
````

## 4. Document simulation setup

### 4.1 Simulation setup $-$ Brio-Wu MHD Shock Tube

- Equations: 1D ideal magnetohydrodynamics in conservative form, with adiabatic index $\gamma=2$

- Domain: $x\in[0,1]$, discretized with $N_{x}=1000$ uniformly spaced finite-volume cells.

- Initial condition: Riemann problem with a discontinuity at $x=0.5$. The left and right states are
$$
\left(\rho,p,v_{x},v_{y},v_{z},B_{x},B_{y},B_{z}\right)_{L}=\left(1.0,1.0,0,0,0,0.75,1.0,0\right)\text{ },
$$
$$
\left(\rho,p,v_{x},v_{y},v_{z},B_{x},B_{y},B_{z}\right)_{R}=\left(0.125,0.1,0,0,0,0.75,-1.0,0\right)\text{ },
$$
which is the canonical Brio–Wu configuration.

- Boundary conditions: (describe what you actually used: typically outflow / zero-gradient.)

- Numerical method: finite-volume scheme with approximate Riemann solver (e.g. HLLD) and high-resolution reconstruction (e.g. MUSCL), advanced in time with a TVD Runge–Kutta method. The CFL condition is chosen so that the scheme remains stable and accurately captures shocks and discontinuities

- Output: The simulation is run up to final time $T_{\text{final}}=2.00$, storing snapshots every $\Delta t=0.05$, for a total of $N_{t}=41$ time steps. At each time we record density $\rho(x,t)$, pressure $p(x,t)$, velocity components $v_{x},v_{y},v_{z}$, magnetic field components $B_{x},B_{y},B_{z}$ and total energy density.
