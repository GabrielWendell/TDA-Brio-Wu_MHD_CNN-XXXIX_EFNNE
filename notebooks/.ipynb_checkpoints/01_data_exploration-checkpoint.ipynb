### 0. Imports and path setup

````python
import os 
import numpy as np
import matplotlib.pyplot as plt

import sys, os
sys.path.append(os.path.abspath("../"))
from src.io.load_brio_wu import load_brio_wu_data

DATA_DIR = "../data/raw"

# Make plots nicer
import smplotlib
def opt_plot():
    plt.grid(True, linestyle = ':', color = '0.50')
    plt.rcParams['text.usetex'] = True
    plt.rcParams['font.size'] = 15
    plt.minorticks_on()
    plt.tick_params(axis = 'both', which = 'minor', direction = "in",
                        top = True, right = True, length = 5, width = 1, labelsize = 15)
    plt.tick_params(axis = 'both', which = 'major', direction = "in",
                        top = True, right = True, length = 8, width = 1, labelsize = 15)
````

### 1. Load the dataset

````python
dataset = load_brio_wu_data(DATA_DIR)

times = dataset.times       # shape (N_t,)
x     = dataset.x           # shape (N_x,)
rho   = dataset.rho         # shape (N_t, N_x)
p     = dataset.p           # shape (N_t, N_x)

print(f"Loaded Brio–Wu dataset from {DATA_DIR}")
print(f"Number of time steps N_t = {times.shape[0]}")
print(f"Number of spatial points N_x = {x.shape[0]}")
print(f"rho shape = {rho.shape}, p shape = {p.shape}")

if dataset.extra:
    print("Extra fields available:", list(dataset.extra.keys()))
````

### 2. Snapshot of density and pressure

````python
# Helper function to plot a single variable at various times

def plot_profiles_at_times(x, field, times, t_query, title = "", ylabel = ""):
    """
    Plot field(x, t) for a list of times t_query.

    Parameters
    ----------
    x : array, shape (N_x,)
        Spatial grid.
    field : array, shape (N_t, N_x)
        Field values: field[t_idx, x_idx].
    times : array, shape (N_t,)
        Time grid.
    t_query : list of float
        Desired times to plot. The nearest time index is used.
    """
    plt.figure(figsize = (8, 5))
    for tq in t_query:
        idx = int(np.argmin(np.abs(times - tq)))
        plt.plot(x, field[idx], label = f"t = {times[idx]:.2f}")
    plt.xlabel("x")
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend(loc = "best")
    plt.tight_layout()
    opt_plot()
    plt.show()
    
# Cell 4: profiles for density and pressure

t_query = [0.0, 0.2, 0.5, 1.0, 2.0]

plot_profiles_at_times(
    x, rho, times, t_query,
    title = "Density profiles in Brio-Wu shock tube",
    ylabel = r"$\rho(x,t)$"
)

plot_profiles_at_times(
    x, p, times, t_query,
    title = "Pressure profiles in Brio-Wu shock tube",
    ylabel = r"$p(x,t)$"
)
````

### 3. Space-time diagrams ($x-t$ maps)

We want a global view of the evolution of the fields, which will be very useful later for linking physics $\iff$ topology.

````python
# Cell 5: space–time diagrams for rho and p

def plot_xt_diagram(x, times, field, title = "", 
                    vmin = None, vmax = None, cmap = "jet"):
    """
    Plot a 2D x–t diagram of a field.

    Parameters
    ----------
    x : array, shape (N_x,)
        Spatial grid.
    times : array, shape (N_t,)
        Time grid.
    field : array, shape (N_t, N_x)
        Field values field[t_idx, x_idx].
    """
    T, Nx = field.shape

    # We need meshgrids for pcolormesh: shape (N_t, N_x)
    X, Tgrid = np.meshgrid(x, times)

    plt.figure(figsize = (8, 4))
    pcm = plt.pcolormesh(
        X, Tgrid, field,
        shading = "auto", cmap = cmap,
        vmin = vmin, vmax = vmax
    )
    plt.colorbar(pcm, label = "Field value")
    plt.xlabel(r"$x$")
    plt.ylabel(r"$t$")
    plt.title(title)
    plt.tight_layout()
    plt.show()

# Density x–t diagram
plot_xt_diagram(
    x, times, rho,
    title = r"Brio-Wu density evolution: $x-t$ diagram"
)

# Pressure x–t diagram
plot_xt_diagram(
    x, times, p,
    title = r"Brio-Wu pressure evolution: $x-t$ diagram"
)
````

If you want to enforce the same color scale across fields or across different runs, you can compute `vmin` and `vmax` explicitly:

```python
rho_vmin, rho_vmax = rho.min(), rho.max()
p_vmin, p_vmax     = p.min(),  p.max()

plot_xt_diagram(x, times, rho, title = r"Density $x-t$ diagram",
                vmin = rho_vmin, vmax = rho_vmax)

plot_xt_diagram(x, times, p, title = r"Pressure $x-t$ diagram",
                vmin = p_vmin, vmax = p_vmax)
```

### 4. Quick consistency diagnostics

Finally, a small cell to verify that your time grid is uniform and matches the theoretical $\Delta t=0.05$:

````python
# Consistency checks

dt = np.diff(times)
print("Δt values:", dt)
print("\nMean Δt:", dt.mean())
print("Max deviation from 0.05:", np.abs(dt - 0.05).max())

# Check conservation of total mass as a quick sanity check
total_mass = np.trapz(rho, x, axis = 1)  # Integrate over x at each time
plt.figure(figsize = (8, 4))
plt.plot(times, total_mass, marker = "o")
plt.xlabel(r"$t$")
plt.ylabel(r"Total mass")
plt.title("Total mass vs time")
plt.tight_layout()
opt_plot()
plt.show()
````

**OBS:** $\text{Total Mass}=\int\rho\text{ d}x$

If the finite-volume scheme is implemented correctly, the total mass should be approximately constant (up to numerical error).
